```python
from sympy.solvers import solve
from sympy import Symbol
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = Symbol('x')
```

```python
f = 0.6*x**3 + 5.5*x**2 +10*x - 5
```

```python
roots = solve(f, x)
```

```python
dif = solve(f.diff(x))
list_x = [i for i in dif]
list_y = [0.6*x**3 + 5.5*x**2 +10*x - 5 for x in list_x]
xmin,ymin = list_x[0],list_y[0]
xmax,ymax = list_x[1],list_y[1]
```

```python
fig , ax = plt.subplots(nrows = 2, ncols =2)

fig.set_facecolor('green')
fig.set_figwidth(20)
fig.set_figheight(15)

x = np.linspace(-7, 1)
y = 0.6*x**3 + 5.5*x**2 + 10*x - 5

#Построение окон
for i in range(2):
    for j in range(2):
        ax[i][j].set_facecolor('yellow')
        ax[i][j].set_xlabel('ось абцис')
        ax[i][j].set_ylabel('ось ординат')
        ax[i][j].set_xlim([-7, 1])
        ax[i][j].set_ylim([-11, 9])
        ax[i][j].plot(x, x*0,color='black')
        ax[i][j].plot(y*0,y,color='black')
        ax[i][j].plot(x,y, color = 'crimson')
        
#График функции        
ax[0][0].set_title(f'График функции f(x)={f}')

#Точки и корни
ax[0][1].set_title('Точки min, max, корни')
ax[0][1].scatter(round(xmin,3),round(ymin,3))
ax[0][1].scatter(round(xmax,3),round(ymax,3))
for i in roots:
    ax[0][1].scatter(round(i, 3), 0, marker = 's', color = 'black')
    
#функция отрицательна
ax[1][0].set_title(f'Функция отрицательна')
ax[1][0].fill_between(x, x*0, y, where = (y < 0), facecolor='blue')

#функция положительна
ax[1][1].set_title(f'Функция положительна')
ax[1][1].fill_between(x, x*0, y, where= (y > 0), facecolor='orange')

plt.show()
```
